```python

# ------------------------------------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------- IMPORTACOES -----------------------------------------------------------------
# ------------------------------------------------------------------------------------------------------------------------------------
import ipywidgets as widgets
from IPython.display import display
from pathlib import Path
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, cohen_kappa_score, matthews_corrcoef
import fastai
from fastai.vision.all import *
from fastai.vision.widgets import *
import shutil  # Novo import para manipulação de arquivos
import torch  # Novo import para exportação ONNX
from google.colab import drive
from collections import Counter
import os
import matplotlib.pyplot as plt
import seaborn as sns
from contextlib import redirect_stdout
from io import StringIO
import numpy as np
import io
from fastai.vision.data import ImageBlock, CategoryBlock
from fastai.data.external import untar_data, URLs
from fastai.data.transforms import get_image_files, parent_label, RandomSplitter
from PIL import Image
import pandas as pd
from fastai.vision import models


# ------------------------------------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------- GLOBAIS -----------------------------------------------------------------
# ------------------------------------------------------------------------------------------------------------------------------------
global_path = None
learn = None
dls_train = None
datablock = None
export_path = '/content/'
resize = None
aug_transforms_size = None

preds = None
targs = None
#model_dict = {
#    'AlexNet': alexnet,
#    'ResNet18': resnet18,
#    'ResNet34': resnet34
#}

model_dict = {
    'AlexNet': AlexNet,
    'alexnet': alexnet,
    'AlexNet_Weights': AlexNet_Weights,
    'ConvNeXt': ConvNeXt,
    'convnext_base': convnext_base,
    'ConvNeXt_Base_Weights': ConvNeXt_Base_Weights,
    'convnext_large': convnext_large,
    'ConvNeXt_Large_Weights': ConvNeXt_Large_Weights,
    'convnext_small': convnext_small,
    'ConvNeXt_Small_Weights': ConvNeXt_Small_Weights,
    'convnext_tiny': convnext_tiny,
    'ConvNeXt_Tiny_Weights': ConvNeXt_Tiny_Weights,
    'DenseNet': DenseNet,
    'densenet121': densenet121,
    'DenseNet121_Weights': DenseNet121_Weights,
    'densenet161': densenet161,
    'DenseNet161_Weights': DenseNet161_Weights,
    'densenet169': densenet169,
    'DenseNet169_Weights': DenseNet169_Weights,
    'densenet201': densenet201,
    'DenseNet201_Weights': DenseNet201_Weights,
    'EfficientNet': EfficientNet,
    'efficientnet_b0': efficientnet_b0,
    'EfficientNet_B0_Weights': EfficientNet_B0_Weights,
    'efficientnet_b1': efficientnet_b1,
    'EfficientNet_B1_Weights': EfficientNet_B1_Weights,
    'efficientnet_b2': efficientnet_b2,
    'EfficientNet_B2_Weights': EfficientNet_B2_Weights,
    'efficientnet_b3': efficientnet_b3,
    'EfficientNet_B3_Weights': EfficientNet_B3_Weights,
    'efficientnet_b4': efficientnet_b4,
    'EfficientNet_B4_Weights': EfficientNet_B4_Weights,
    'efficientnet_b5': efficientnet_b5,
    'EfficientNet_B5_Weights': EfficientNet_B5_Weights,
    'efficientnet_b6': efficientnet_b6,
    'EfficientNet_B6_Weights': EfficientNet_B6_Weights,
    'efficientnet_b7': efficientnet_b7,
    'EfficientNet_B7_Weights': EfficientNet_B7_Weights,
    'efficientnet_v2_l': efficientnet_v2_l,
    'EfficientNet_V2_L_Weights': EfficientNet_V2_L_Weights,
    'efficientnet_v2_m': efficientnet_v2_m,
    'EfficientNet_V2_M_Weights': EfficientNet_V2_M_Weights,
    'efficientnet_v2_s': efficientnet_v2_s,
    'EfficientNet_V2_S_Weights': EfficientNet_V2_S_Weights,
    'GoogLeNet': GoogLeNet,
    'googlenet': googlenet,
    'GoogLeNet_Weights': GoogLeNet_Weights,
    'GoogLeNetOutputs': GoogLeNetOutputs,
    'Inception3': Inception3,
    'inception_v3': inception_v3,
    'Inception_V3_Weights': Inception_V3_Weights,
    'InceptionOutputs': InceptionOutputs,
    'MaxVit': MaxVit,
    'maxvit_t': maxvit_t,
    'MaxVit_T_Weights': MaxVit_T_Weights,
    'MNASNet': MNASNet,
    'mnasnet0_5': mnasnet0_5,
    'MNASNet0_5_Weights': MNASNet0_5_Weights,
    'mnasnet0_75': mnasnet0_75,
    'MNASNet0_75_Weights': MNASNet0_75_Weights,
    'mnasnet1_0': mnasnet1_0,
    'MNASNet1_0_Weights': MNASNet1_0_Weights,
    'mnasnet1_3': mnasnet1_3,
    'MNASNet1_3_Weights': MNASNet1_3_Weights,
    'mobilenet_v2': mobilenet_v2,
    'MobileNet_V2_Weights': MobileNet_V2_Weights,
    'mobilenet_v3_large': mobilenet_v3_large,
    'MobileNet_V3_Large_Weights': MobileNet_V3_Large_Weights,
    'mobilenet_v3_small': mobilenet_v3_small,
    'MobileNet_V3_Small_Weights': MobileNet_V3_Small_Weights,
    'MobileNetV2': MobileNetV2,
    'MobileNetV3': MobileNetV3,
    'RegNet': RegNet,
    'regnet_x_16gf': regnet_x_16gf,
    'RegNet_X_16GF_Weights': RegNet_X_16GF_Weights,
    'regnet_x_1_6gf': regnet_x_1_6gf,
    'RegNet_X_1_6GF_Weights': RegNet_X_1_6GF_Weights,
    'regnet_x_32gf': regnet_x_32gf,
    'RegNet_X_32GF_Weights': RegNet_X_32GF_Weights,
    'regnet_x_3_2gf': regnet_x_3_2gf,
    'RegNet_X_3_2GF_Weights': RegNet_X_3_2GF_Weights,
    'regnet_x_400mf': regnet_x_400mf,
    'RegNet_X_400MF_Weights': RegNet_X_400MF_Weights,
    'regnet_x_800mf': regnet_x_800mf,
    'RegNet_X_800MF_Weights': RegNet_X_800MF_Weights,
    'regnet_x_8gf': regnet_x_8gf,
    'RegNet_X_8GF_Weights': RegNet_X_8GF_Weights,
    'regnet_y_128gf': regnet_y_128gf,
    'RegNet_Y_128GF_Weights': RegNet_Y_128GF_Weights,
    'regnet_y_16gf': regnet_y_16gf,
    'RegNet_Y_16GF_Weights': RegNet_Y_16GF_Weights,
    'regnet_y_1_6gf': regnet_y_1_6gf,
    'RegNet_Y_1_6GF_Weights': RegNet_Y_1_6GF_Weights,
    'regnet_y_32gf': regnet_y_32gf,
    'RegNet_Y_32GF_Weights': RegNet_Y_32GF_Weights,
    'regnet_y_3_2gf': regnet_y_3_2gf,
    'RegNet_Y_3_2GF_Weights': RegNet_Y_3_2GF_Weights,
    'regnet_y_400mf': regnet_y_400mf,
    'RegNet_Y_400MF_Weights': RegNet_Y_400MF_Weights,
    'regnet_y_800mf': regnet_y_800mf,
    'RegNet_Y_800MF_Weights': RegNet_Y_800MF_Weights,
    'regnet_y_8gf': regnet_y_8gf,
    'RegNet_Y_8GF_Weights': RegNet_Y_8GF_Weights,
    'ResNet': ResNet,
    'resnet101': resnet101,
    'ResNet101_Weights': ResNet101_Weights,
    'resnet152': resnet152,
    'ResNet152_Weights': ResNet152_Weights,
    'resnet18': resnet18,
    'ResNet18_Weights': ResNet18_Weights,
    'resnet34': resnet34,
    'ResNet34_Weights': ResNet34_Weights,
    'resnet50': resnet50,
    'ResNet50_Weights': ResNet50_Weights,
    'resnext101_32x8d': resnext101_32x8d,
    'ResNeXt101_32X8D_Weights': ResNeXt101_32X8D_Weights,
    'resnext101_64x4d': resnext101_64x4d,
    'ResNeXt101_64X4D_Weights': ResNeXt101_64X4D_Weights,
    'resnext50_32x4d': resnext50_32x4d,
    'ResNeXt50_32X4D_Weights': ResNeXt50_32X4D_Weights,
    'shufflenet_v2_x0_5': shufflenet_v2_x0_5,
    'ShuffleNet_V2_X0_5_Weights': ShuffleNet_V2_X0_5_Weights,
    'shufflenet_v2_x1_0': shufflenet_v2_x1_0,
    'ShuffleNet_V2_X1_0_Weights': ShuffleNet_V2_X1_0_Weights,
    'shufflenet_v2_x1_5': shufflenet_v2_x1_5,
    'ShuffleNet_V2_X1_5_Weights': ShuffleNet_V2_X1_5_Weights,
    'shufflenet_v2_x2_0': shufflenet_v2_x2_0,
    'ShuffleNet_V2_X2_0_Weights': ShuffleNet_V2_X2_0_Weights,
    'ShuffleNetV2': ShuffleNetV2,
    'SqueezeNet': SqueezeNet,
    'squeezenet1_0': squeezenet1_0,
    'SqueezeNet1_0_Weights': SqueezeNet1_0_Weights,
    'squeezenet1_1': squeezenet1_1,
    'SqueezeNet1_1_Weights': SqueezeNet1_1_Weights,
    'swin_b': swin_b,
    'Swin_B_Weights': Swin_B_Weights,
    'swin_s': swin_s,
    'Swin_S_Weights': Swin_S_Weights,
    'swin_t': swin_t,
    'Swin_T_Weights': Swin_T_Weights,
    'swin_v2_b': swin_v2_b,
    'Swin_V2_B_Weights': Swin_V2_B_Weights,
    'swin_v2_s': swin_v2_s,
    'Swin_V2_S_Weights': Swin_V2_S_Weights,
    'swin_v2_t': swin_v2_t,
    'Swin_V2_T_Weights': Swin_V2_T_Weights,
    'SwinTransformer': SwinTransformer,
    'VGG': VGG,
    'vgg11': vgg11,
    'vgg11_bn': vgg11_bn,
    'VGG11_BN_Weights': VGG11_BN_Weights,
    'VGG11_Weights': VGG11_Weights,
    'vgg13': vgg13,
    'vgg13_bn': vgg13_bn,
    'VGG13_BN_Weights': VGG13_BN_Weights,
    'VGG13_Weights': VGG13_Weights,
    'vgg16': vgg16,
    'vgg16_bn': vgg16_bn,
    'VGG16_BN_Weights': VGG16_BN_Weights,
    'VGG16_Weights': VGG16_Weights,
    'vgg19': vgg19,
    'vgg19_bn': vgg19_bn,
    'VGG19_BN_Weights': VGG19_BN_Weights,
    'VGG19_Weights': VGG19_Weights,
    'VisionTransformer': VisionTransformer,
    'vit_b_16': vit_b_16,
    'ViT_B_16_Weights': ViT_B_16_Weights,
    'vit_b_32': vit_b_32,
    'ViT_B_32_Weights': ViT_B_32_Weights,
    'vit_h_14': vit_h_14,
    'ViT_H_14_Weights': ViT_H_14_Weights,
    'vit_l_16': vit_l_16,
    'ViT_L_16_Weights': ViT_L_16_Weights,
    'vit_l_32': vit_l_32,
    'ViT_L_32_Weights': ViT_L_32_Weights,
    'wide_resnet101_2': wide_resnet101_2,
    'Wide_ResNet101_2_Weights': Wide_ResNet101_2_Weights,
    'wide_resnet50_2': wide_resnet50_2,
    'Wide_ResNet50_2_Weights': Wide_ResNet50_2_Weights
}



output_preparation = widgets.Output()
output_training = widgets.Output()
output_evaluation = widgets.Output()
output_fine_tuning = widgets.Output()
output_prediction = widgets.Output()
output_export = widgets.Output()
output_insert_test_path = widgets.Output()
output_test_metrics = widgets.Output()
global_evaluation_result = "" # Variável global para armazenar os resultados finais da avaliação
global_final_result = "" # Variável global para armazenar os resultados finais




# ------------------------------------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------- FUNCOES-----------------------------------------------------------------
# ------------------------------------------------------------------------------------------------------------------------------------
# ------------------ 0-PREPARAÇÃO DOS DADOS -----------------

# Funções para as abas
def insert_path(b):
  with output_preparation:
    global global_path
    drive.mount('/content/gdrive', force_remount=True)
    path = Path(f'/content/{train_path_text.value}')

    if path.exists():
        global_path = path
        print(f"Caminho '{train_path_text.value}' inserido com sucesso.")
    else:
        print(f"Erro: O caminho '{train_path_text.value}' não foi encontrado.")

# Função para gerar o extrato do dataset
def generate_dataset_extract(b):
    global global_path  # Usando o path global já definido no código anterior

    with output_preparation:
        # Verificando se o caminho global está definido
        if global_path is not None:

            class_counter = Counter()
            extension_counter = Counter()
            total_images = 0  # Contador para o número total de imagens

            # Loop para contar as classes e as extensões dos arquivos
            for subdir, _, files in os.walk(global_path):
                for file_name in files:
                    total_images += 1  # Incrementar a contagem total de imagens

                    # Contar classes
                    class_name = os.path.basename(subdir)
                    class_counter[class_name] += 1

                    # Contar extensões
                    file_extension = os.path.splitext(file_name)[-1]
                    extension_counter[file_extension] += 1

            # Imprimir os resultados
            print("Contagem de classes:", class_counter)
            print("Contagem de extensões:", extension_counter)
            print("Contagem total de imagens:", total_images)

            # Plotting
            fig, axes = plt.subplots(1, 2, figsize=(11, 4))

            # Plot para a distribuição de classes
            sns.barplot(x=list(class_counter.keys()), y=list(class_counter.values()), palette='Blues', ax=axes[0])
            axes[0].set_title('Distribuição de classes no conjunto de treinamento')
            axes[0].set_xlabel('Classe')
            axes[0].set_ylabel('Numero de amostras')
            for i, v in enumerate(class_counter.values()):
                axes[0].text(i, v, str(v), color='black', ha="center")

            # Plot para a distribuição de extensões de arquivo
            sns.barplot(x=list(extension_counter.keys()), y=list(extension_counter.values()), palette='Blues', ax=axes[1])
            axes[1].set_title('Distribuição das extensões de arquivo no dataset')
            axes[1].set_xlabel('Extensões de arquivos')
            axes[1].set_ylabel('Numero de arquivos')
            for i, v in enumerate(extension_counter.values()):
                axes[1].text(i, v, str(v), color='black', ha="center")

            plt.tight_layout()
            plt.show()

        else:
            print("Por favor, insira o caminho do conjunto de treinamento primeiro.")

# Função preparacao de dados
def prepare_data(b):
    global global_path, dls_train, datablock, resize, aug_transforms_size  # datablock, resize, e aug_transforms_size adicionados como globais

    # Limpa a saída anterior (comentado para manter as mensagens anteriores)
    # output_preparation.clear_output(wait=True)

    with output_preparation:
        resize = int(resize_text.value)
        aug_transforms_size = int(aug_transforms_text.value)

        datablock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                              get_items=get_image_files,
                              splitter=RandomSplitter(seed=42),
                              get_y=parent_label,
                              item_tfms=Resize(resize),
                              batch_tfms=aug_transforms(size=aug_transforms_size))

        dls_train = datablock.dataloaders(global_path)

        print(f"Preparação dos dados concluída com sucesso. Resize: {resize}, Aug Transforms Size: {aug_transforms_size}")




# ------------------ 1-TREINAMENTO -----------------

def start_training(b):
    final_metrics = {}  # Inicialize um dicionário para armazenar as métricas finais

    with output_training:
        print("Aguarde... iniciando o treinamento.")

        global dls_train, learn
        selected_model = model_dropdown.value
        epochs = int(epochs_text.value)
        lr = float(lr_text.value)

        # Usar o modelo selecionado aqui
        selected_model_class = model_dict[selected_model]
        learn = vision_learner(dls_train, selected_model_class, metrics=accuracy)

        # Realizar o treinamento e armazenar métricas finais
        learn.fit_one_cycle(epochs, lr)

        final_metrics['Training Loss'] = learn.recorder.final_record[0]
        final_metrics['Validation Loss'] = learn.recorder.final_record[1]
        final_metrics['Validation Accuracy'] = learn.recorder.final_record[2] * 100  # Multiplicado por 100 para converter de proporção para porcentagem

        learn.save('modelo')

    # Limpar saída e imprimir métricas finais
    with output_training:
        output_training.clear_output(wait=True)
        print("Treinamento e salvamento do modelo concluídos.")
        for metric, value in final_metrics.items():
            print(f"{metric}: {value:.4f}")

        # Verifique se os dados estão disponíveis
        if len(learn.recorder.values) > 0 and len(learn.recorder.values[0]) >= 3:  # Verifica se há dados e se a acurácia é uma das métricas
            # Transforma a lista de listas em um array NumPy
            values_array = np.array(learn.recorder.values)

            # Obtém a época com a melhor acurácia de validação
            best_epoch = np.argmax(values_array[:, 2])  # supondo que a acurácia é a terceira métrica na lista
            print(f"Melhor acurácia de validação alcançada na época: {best_epoch + 1}")  # Adiciona 1 porque a contagem de épocas começa de 1
        else:
            print("Não foi possível determinar a melhor época para acurácia de validação.")

# ------------------ 2-AVALIAÇÃO DO MODELO -----------------

def show_train_acc(b):
    global global_evaluation_result

    with output_evaluation:
        print("Calculando acurácia de treinamento...")
        global learn
        preds, targs = learn.get_preds(dl=dls_train.train)
        accuracy = accuracy_score(targs, preds.argmax(dim=-1))

        global_evaluation_result += f"Calculando acurácia de treinamento...\n"
        global_evaluation_result += f"Acurácia de Treinamento: {accuracy * 100:.2f}%\n"

        output_evaluation.clear_output(wait=True)
        print(global_evaluation_result)

def show_val_metrics(b):
    global global_evaluation_result

    with output_evaluation:
        print("Calculando métricas de validação...")
        global learn
        interp = ClassificationInterpretation.from_learner(learn)

        # Capture a saída da função
        with io.StringIO() as buf, redirect_stdout(buf):
            interp.print_classification_report()
            report = buf.getvalue()

        global_evaluation_result += "Calculando métricas de validação...\n"
        global_evaluation_result += report

        output_evaluation.clear_output(wait=True)
        print(global_evaluation_result)

from IPython.display import display, clear_output
import matplotlib.pyplot as plt

from IPython.display import display, clear_output
from ipywidgets import Image
import io

def show_confusion_matrix(b):
    global global_evaluation_result

    with output_evaluation:
        print("Gerando matriz de confusão...")
        global learn
        interp = ClassificationInterpretation.from_learner(learn)

        global_evaluation_result += "Gerando matriz de confusão...\n"

        output_evaluation.clear_output(wait=True)
        print(global_evaluation_result)

        # Captura a figura da matriz de confusão em um buffer
        interp.plot_confusion_matrix()
        buf = io.BytesIO()
        plt.savefig(buf, format='png')
        plt.close()  # Feche a figura para evitar uma exibição duplicada

        # Exibe a figura capturada dentro do widget de saída
        image_widget = Image(value=buf.getvalue(), format='png')
        display(image_widget)




# ------------------ 3-FINE TUNING -----------------

# Função para buscar o melhor Learning Rate
def find_best_lr(b):
    global global_final_result

    with output_fine_tuning:
        # Seu código para encontrar o melhor Learning Rate aqui
        learn.unfreeze()
        lr_min, lr_steep, lr_valley, lr_slide = learn.lr_find(suggest_funcs=(minimum, steep, valley, slide))

        global_final_result += "Buscando o melhor Learning Rate. Aguarde...\n"
        global_final_result += f"lr_min = {lr_min}\n"
        global_final_result += f"lr_steep = {lr_steep}\n"
        global_final_result += f"lr_valley = {lr_valley}\n"
        global_final_result += f"lr_slide = {lr_slide}\n"

        # Limpar saída e imprimir o resultado final
        output_fine_tuning.clear_output(wait=True)
        print(global_final_result)







# Função para iniciar o Fine Tuning
def start_fine_tuning(b):
    global global_final_result
    final_metrics = {}  # Inicialize um dicionário para armazenar as métricas finais

    with output_fine_tuning:
        # Seu código para Fine Tuning aqui
        epochs = int(fine_tune_epochs_text.value)
        lr = float(lr_text.value)  # Supondo que você quer usar o mesmo lr definido anteriormente

        learn.fine_tune(epochs, lr)
        learn.save('modelo')

        # Obter métricas finais
        if len(learn.recorder.values) > 0:
            final_epoch_metrics = learn.recorder.values[-1]  # As métricas da última época
            final_metrics['Training Loss (Fine Tuning)'] = final_epoch_metrics[0]
            final_metrics['Validation Loss (Fine Tuning)'] = final_epoch_metrics[1]
            final_metrics['Validation Accuracy (Fine Tuning)'] = final_epoch_metrics[2] * 100  # Convertido para porcentagem

        global_final_result += "Iniciando Fine Tuning. Aguarde...\n"
        global_final_result += "Otimização e salvamento do modelo concluídos.\n"

        # Limpar saída e imprimir o resultado final e as métricas
        output_fine_tuning.clear_output(wait=True)
        print(global_final_result)
        for metric, value in final_metrics.items():
            print(f"{metric}: {value:.4f}")

# ------------------ 4-PREDIÇÃO -----------------

def insert_test_path(b):
    global path_test, learn  # Variáveis globais
    path_test = Path(f'/content/{test_path_text.value}')

    with output_prediction:   # Todas as saídas devem estar dentro deste bloco
        print(f"Tentando carregar o modelo a partir de '/content/{test_path_text.value}'...")

        # Tente carregar o modelo e capture qualquer exceção que possa ocorrer
        try:
            learn.load('modelo')
            print("Modelo carregado com sucesso.")
        except Exception as e:
            print(f"Erro ao carregar o modelo: {e}")
            return

        if path_test.exists():
            print(f"Caminho '{test_path_text.value}' existe.")
        else:
            print(f"Erro: O caminho '{test_path_text.value}' não foi encontrado.")





# Função para iniciar a predição
def start_prediction(b):
    global output_prediction, path_test, learn, resize, aug_transforms_size  # Variáveis globais
    global preds, targs, decoded_preds, decoded_targs, test_dls  # tornar as variáveis globais

    with output_prediction:  # Adicione esta linha
        try:

            print("Iniciando predição. Aguarde...")

            # Crie um DataBlock para as imagens de teste
            test_datablock = DataBlock(
                blocks=(ImageBlock, CategoryBlock),
                get_items=get_image_files,
                splitter=RandomSplitter(valid_pct=0, seed=42),
                get_y=parent_label,
                item_tfms=Resize(resize),
                batch_tfms=aug_transforms(size=aug_transforms_size)
            )

            print("DataBlock criado.")

            # Crie um DataLoader para as imagens de teste
            test_dls = test_datablock.dataloaders(path_test, bs=64)

            print("DataLoader criado.")

            # Verifique se o modelo foi inicializado
            if learn is not None:
                # Faça as previsões
                preds, targs = learn.get_preds(dl=test_dls.train)

                print("Previsões obtidas.")

                # Decodificar as previsões e as verdadeiras etiquetas para obter as classes
                decoded_preds = [test_dls.vocab[i] for i in preds.argmax(dim=-1)]
                decoded_targs = [test_dls.vocab[i] for i in targs]

                #print(decoded_preds)
                print("Predição concluída.")
            else:
                print("O modelo não foi inicializado.")

        except Exception as e:
            print(f"Erro durante a predição: {e}")




def show_test_metrics(b):
    with output_prediction:  # Altere para output_prediction se você quiser a mesma aba de start_prediction
        print("Função show_test_metrics foi chamada.")
        try:
            # Dentro do bloco try
            global preds, targs, decoded_preds, decoded_targs, test_dls  # Variáveis globais

            # Converta as etiquetas verdadeiras para suas formas decodificadas (nomes das classes)
            decoded_targs = [test_dls.vocab[i] for i in targs]

            # Calcular a acurácia
            acc = accuracy_score(decoded_targs, decoded_preds)
            print(f"Acurácia: {acc * 100:.2f}%")

            # Calcular outras métricas (precisão, recall, F1-score)
            report = classification_report(decoded_targs, decoded_preds)
            print("Relatório de Classificação:")
            print(report)

            # Calcule a matriz de confusão
            cm = confusion_matrix(decoded_targs, decoded_preds)

            # Inicialize um array para armazenar a especificidade para cada classe
            specificities = []

            # Calcule a especificidade para cada classe
            for i in range(len(cm)):
                tn = np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]
                fp = np.sum(cm[:, i]) - cm[i, i]
                specificity = tn / (tn + fp)
                specificities.append(specificity)

            # Imprima as especificidades
            for i, specificity in enumerate(specificities):
                print(f"Especificidade para a classe {i}: {specificity:.4f}")

            # Calculando Cohen's Kappa e Matthews Correlation Coefficient
            kappa = cohen_kappa_score(decoded_targs, decoded_preds)
            mcc = matthews_corrcoef(decoded_targs, decoded_preds)

            print(f"Cohen's Kappa: {kappa}")
            print(f"Matthews Correlation Coefficient: {mcc}")

        except Exception as e:
            print(f"Erro ao calcular métricas: {e}")


# Função para mostrar a matriz de confusão do conjunto de teste
def show_test_confusion_matrix(b):
  with output_prediction:    # Adicione esta linha
    try:
        global decoded_targs, decoded_preds, test_dls  # Variáveis globais

        # Calcular a matriz de confusão
        cm = confusion_matrix(decoded_targs, decoded_preds)

        # Transformar em um DataFrame para facilitar a visualização
        class_names = test_dls.vocab
        df_cm = pd.DataFrame(cm, index=class_names, columns=class_names)

        # Criar um heatmap
        plt.figure(figsize=(10, 8))
        sns.heatmap(df_cm, annot=True, fmt="d", cmap='Blues')
        plt.title("Matriz de Confusão")
        plt.xlabel("Predição")
        plt.ylabel("Verdadeiro")
        plt.show()

    except Exception as e:
        print(f"Erro ao mostrar a matriz de confusão: {e}")

# ------------------ 5-EXPORTAÇÃO -----------------

# Função para atualizar o caminho de exportação
def update_export_path(b):
    global export_path
    export_path = export_path_text.value
    print(f"Caminho de exportação atualizado para {export_path}")

# Funções para exportar modelos
def export_pth(b):
    if learn is not None:
      with output_export:
        file_path = f"{export_path}/modelo_exportado.pth"
        learn.save(file_path)
        print(f"Modelo .pth exportado com sucesso para {file_path}")
    else:
        print("Erro: Nenhum modelo disponível para exportação.")

def export_pkl(b):
    if learn is not None:
      with output_export:
        file_path = f"{export_path}/modelo_exportado.pkl"
        learn.export(file_path)
        print(f"Modelo .pkl exportado com sucesso para {file_path}")
    else:
        print("Erro: Nenhum modelo disponível para exportação.")

def export_onnx(b):
    if learn is not None:
      with output_export:
        file_path = f"{export_path}/modelo_exportado.onnx"
        x = torch.randn(1, 3, 224, 224).cuda()  # Mover o tensor para a GPU
        learn.model.eval()
        torch.onnx.export(learn.model, x, file_path, export_params=True, opset_version=10, do_constant_folding=True, input_names=['input'], output_names=['output'])
        print(f"Modelo ONNX exportado com sucesso para {file_path}")
    else:
        print("Erro: Nenhum modelo disponível para exportação.")

# ------------------------------------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------- WIDGETS -----------------------------------------------------------------
# ------------------------------------------------------------------------------------------------------------------------------------
# ------------------ WIDGET LAYOUTS -----------------
text_layout = widgets.Layout(width='300px')
button_layout = widgets.Layout(width='200px')

# ------------------ 0-PREPARAÇÃO DOS DADOS -----------------
# ------ TRAINING PATH -----
train_path_text = widgets.Text(value='', layout=text_layout)
train_path_label = widgets.Label('Entrada das imagens de treinamento: Ex: /gdrive/Shared drives/... ou /gdrive/My Drive/...', layout=widgets.Layout(width='600px'))
train_path_box = widgets.HBox([train_path_text, train_path_label])
insert_path_button = widgets.Button(description='Inserir Caminho', layout=button_layout)
insert_path_button.on_click(insert_path)

# ------ RESIZE -----
resize_text = widgets.Text(value='460', layout=text_layout)
resize_label = widgets.Label('Resize:', layout=widgets.Layout(width='100px'))
resize_box = widgets.HBox([resize_text, resize_label])

# ------ AUG TRANSFORMS -----
aug_transforms_text = widgets.Text(value='224', layout=text_layout)
aug_transforms_label = widgets.Label('Aug Transforms:', layout=widgets.Layout(width='200px'))
aug_transforms_box = widgets.HBox([aug_transforms_text, aug_transforms_label])

# ------ DATASET -----
dataset_extract_button = widgets.Button(description='Extrato do Dataset', layout=button_layout)
dataset_extract_button.on_click(generate_dataset_extract)
prepare_button = widgets.Button(description='Preparação OK', layout=button_layout)
prepare_button.on_click(prepare_data)

# ------------------ 1-TREINAMENTO -----------------
# ------ MODEL SELECTION -----
#model_list = ['AlexNet', 'ResNet18', 'ResNet32']
model_list = [key for key in model_dict.keys()]
model_dropdown = widgets.Dropdown(options=model_list, value='AlexNet', layout=text_layout)
model_label = widgets.Label('Selecione os modelos disponíveis:', layout=widgets.Layout(width='300px'))
model_box = widgets.HBox([model_dropdown, model_label])

# ------ EPOCHS -----
epochs_text = widgets.Text(value='30', layout=text_layout)
epochs_label = widgets.Label('Épocas:', layout=widgets.Layout(width='100px'))
epochs_box = widgets.HBox([epochs_text, epochs_label])

# ------ LEARNING RATE -----
lr_text = widgets.Text(value='0.001', layout=text_layout)
lr_label = widgets.Label('Taxa de Aprendizado:', layout=widgets.Layout(width='200px'))
lr_box = widgets.HBox([lr_text, lr_label])

# ------ TRAINING BUTTONS -----
train_button = widgets.Button(description='Iniciar Treinamento', layout=button_layout)
train_button.on_click(start_training)
train_acc_button = widgets.Button(description='Acurácia do Treinamento', layout=button_layout)
train_acc_button.on_click(show_train_acc)

# ------------------ 2-AVALIAÇÃO DO MODELO -----------------
# ------ VALIDATION METRICS -----
val_metrics_button = widgets.Button(description='Métricas do Modelo', layout=button_layout)
val_metrics_button.on_click(show_val_metrics)
val_metrics_label = widgets.Label('da Validação', layout=widgets.Layout(width='100px'))
val_metrics_box = widgets.HBox([val_metrics_button, val_metrics_label])
confusion_matrix_button = widgets.Button(description='Matriz de Confusão', layout=button_layout)
confusion_matrix_button.on_click(show_confusion_matrix)

# ------------------ 3-FINE TUNING -----------------

# ------ FINE TUNING -----

# Botão para buscar o melhor Learning Rate
find_lr_button = widgets.Button(description='Buscar melhor LR', layout=button_layout)
find_lr_button.on_click(find_best_lr)
find_lr_label = widgets.Label('Iniciar busca do melhor Learning Rate', layout=widgets.Layout(width='300px'))

# Campo de texto para o número de épocas para fine-tuning
fine_tune_epochs_text = widgets.Text(value='1', layout=text_layout)
fine_tune_epochs_label = widgets.Label('Número de Épocas:', layout=widgets.Layout(width='200px'))

# Botão para iniciar o Fine Tuning
start_fine_tune_button = widgets.Button(description='Iniciar', layout=button_layout)
start_fine_tune_button.on_click(start_fine_tuning)
fine_tune_label = widgets.Label('Iniciar Fine Tuning', layout=widgets.Layout(width='200px'))


# ------------------ 4-PREDIÇÃO -----------------
# ------ TESTING -----
test_path_text = widgets.Text(value='', layout=text_layout)
start_prediction_button = widgets.Button(description='Iniciar Predição', layout=button_layout)
start_prediction_button.on_click(start_prediction)
test_path_label = widgets.Label('\nEntrada das imagens de teste: Ex: /gdrive/Shared drives/... ou /gdrive/My Drive/...', layout=widgets.Layout(width='600px'))
test_path_box = widgets.HBox([test_path_text, test_path_label])
insert_test_path_button = widgets.Button(description='Inserir Caminho', layout=button_layout)
insert_test_path_button.on_click(insert_test_path)
test_metrics_button = widgets.Button(description='Métricas do Teste', layout=button_layout)
test_metrics_button.on_click(show_test_metrics)
test_confusion_matrix_button = widgets.Button(description='Matriz de Confusão', layout=button_layout)
test_confusion_matrix_button.on_click(show_test_confusion_matrix)


# ------------------ 5-EXPORTAÇÃO -----------------
# ------ EXPORT -----
export_path_text = widgets.Text(value=export_path, layout=text_layout)
update_export_path_button = widgets.Button(description='Atualizar Caminho', layout=button_layout)
update_export_path_button.on_click(update_export_path)
export_pth_button = widgets.Button(description='Exportar modelo .pth', layout=button_layout)
export_pth_button.on_click(export_pth)
export_pkl_button = widgets.Button(description='Exportar modelo .pkl', layout=button_layout)
export_pkl_button.on_click(export_pkl)
export_onnx_button = widgets.Button(description='Exportar modelo ONNX', layout=button_layout)
export_onnx_button.on_click(export_onnx)


# ------------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------GUI LAYOUT -----------------------------------------------------------------
# ------------------------------------------------------------------------------------------------------------------------------------

# ------------------ WIDGET LAYOUTS -----------------
text_layout = widgets.Layout(width='300px')
button_layout = widgets.Layout(width='200px')

# ------------------ 0-PREPARAÇÃO DOS DADOS -----------------
# ------ TRAINING PATH -----
preparation_tab = widgets.VBox([
    train_path_box,
    insert_path_button,
    resize_box,
    aug_transforms_box,
    prepare_button,
    output_preparation
])
# Armazena os elementos existentes em uma lista temporária
temp_children_list = list(preparation_tab.children)

# Insere o novo botão na posição desejada (neste caso, posição 2, logo após o botão "Inserir Caminho")
temp_children_list.insert(2, dataset_extract_button)

# Atualiza os children da aba de Preparação de Dados
preparation_tab.children = tuple(temp_children_list)

# ------------------ 1-TREINAMENTO -----------------
# ------ MODEL SELECTION, EPOCHS, LEARNING RATE -----
training_tab = widgets.VBox([
    model_box,
    epochs_box,
    lr_box,
    train_button,
    output_training  # Added output widget
])

# ------------------ 2-AVALIAÇÃO DO MODELO -----------------
# ------ VALIDATION METRICS -----
evaluation_tab = widgets.VBox([
    train_acc_button,
    val_metrics_box,
    confusion_matrix_button,
    output_evaluation  # Added output widget
])

# ------------------ 3-FINE TUNING -----------------
# ------ FINE TUNING OPTIONS -----
fine_tuning_tab = widgets.VBox([
    widgets.HBox([find_lr_button, find_lr_label]),
    widgets.HBox([fine_tune_epochs_text, fine_tune_epochs_label]),
    widgets.HBox([start_fine_tune_button, fine_tune_label]),
    output_fine_tuning
])

# ------------------ 4-PREDIÇÃO -----------------
# ------ TESTING -----
prediction_tab = widgets.VBox([
    test_path_box,
    insert_test_path_button,
    start_prediction_button,
    test_metrics_button,
    test_confusion_matrix_button,
    output_prediction,  # Saída para a predição
    output_insert_test_path,  # Saída para a inserção do caminho de teste
    output_test_metrics  # Saída para as métricas de teste
])



# ------------------ 5-EXPORTAÇÃO -----------------
# ------ EXPORT OPTIONS -----
export_tab = widgets.VBox([
    widgets.HBox([export_path_text, update_export_path_button]),
    export_pth_button,
    export_pkl_button,
    export_onnx_button,
    output_export  # Added output widget
])

# ------------------ MAIN TAB -----------------
tab = widgets.Tab(children=[
    preparation_tab,
    training_tab,
    evaluation_tab,
    fine_tuning_tab,
    prediction_tab,
    export_tab
])
tab.set_title(0, 'Preparação de Dados')
tab.set_title(1, 'Treinamento')
tab.set_title(2, 'Avaliação do Modelo')
tab.set_title(3, 'Fine Tuning')
tab.set_title(4, 'Predição')
tab.set_title(5, 'Exportação')

display(tab)
```
